```python
# Amihud_indicator calculation
# 1. Clean up column names
df3.columns = df3.columns.str.strip()

# 2. Calculate price return between date_x and date_y
df3['price_return'] = (df3['price_y'] - df3['price_x']) / df3['price_x']

# 3. Calculate volume ratio (volume_y / volume_x)
df3['volume_ratio'] = df3['24h_volume_y'] / df3['24h_volume_x']

# 4. Calculate Amihud Illiquidity for each date
# Note: Using absolute % return / volume as proxy
df3['amihud_x'] = np.abs(df3['24h_x']) / df3['24h_volume_x']
df3['amihud_y'] = np.abs(df3['24h_y']) / df3['24h_volume_y']

# 5. Calculate change in Amihud between the two dates
df3['amihud_change'] = df3['amihud_y'] - df3['amihud_x']

# 6. Now print the result
print(df3[['coin', 'price_x', 'price_y', 'price_return', 'volume_ratio', 'amihud_x', 'amihud_y', 'amihud_change']].head())
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Assuming df5 is a DataFrame that contains a 'price_x' column

# Log-transform to stabilize the variance
log_price = np.log(df5['price_x'])

# Fit the ARIMA model (order=(p,d,q) = (5,1,0) in this case)
model = ARIMA(log_price, order=(5, 1, 0))
model_fit = model.fit()

# Forecast the next 30 time steps
log_forecast = model_fit.forecast(steps=30)

# Convert forecast back to original scale
forecast = np.exp(log_forecast)

# Display or return the forecast
print(forecast)

```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Feature engineering
df3['return_1d'] = df3['price_x'].pct_change()
df3['rolling_mean'] = df3['price_x'].rolling(7).mean()
df3['rolling_std'] = df3['price_x'].rolling(7).std()
df3['target'] = df3['price_x'].shift(-1)  # Predict next dayâ€™s price

# Drop missing values due to rolling and shift operations
df = df3.dropna()

# Define features and target
X = df[['return_1d', 'rolling_mean', 'rolling_std']]
y = df['target']

# Split data (no shuffle for time series)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Model training
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Prediction and evaluation
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from scipy.stats import randint

# Define hyperparameter distributions
param_dist = {
    'n_estimators': randint(50, 300),
    'max_depth': randint(5, 50),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5),
    'max_features': ['auto', 'sqrt', 'log2']
}

# Set up the RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_distributions=param_dist,
    n_iter=50,             # Number of combinations to try
    cv=3,                  # 3-fold cross-validation
    n_jobs=-1,             # Use all CPU cores
    scoring='neg_root_mean_squared_error',
    random_state=42
)

# Fit the random search model
random_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", random_search.best_params_)
print("Best RMSE (CV, negative):", random_search.best_score_)
```

```python
# Train model using the best hyperparameters
best_model = RandomForestRegressor(**random_search.best_params_, random_state=42)
best_model.fit(X_train, y_train)

# Predict on test set
y_pred = best_model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Final RMSE on Test Set:", rmse)
```
